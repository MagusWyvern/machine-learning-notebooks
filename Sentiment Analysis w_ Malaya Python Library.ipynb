**Sentiment Analysis w/ Malaya Python Library**

Using the Malaya Library, this model is able to analyze any given text and determine it's sentiment rating. The model will give the negative / neutral / positive scores for the text given.

------------------------------



```python
# We will install the malaya library package from https://github.com/huseinzol05/malaya
!pip3 install malaya

# Import the Malaya package, Tensorflow and PrettyPrint from Python
import malaya
import tensorflow as tf
from pprint import pprint

# Check the loaded Tensorflow version
tf.__version__
```

```python
xlnet = malaya.sentiment.transformer('xlnet')
tiny_albert = malaya.sentiment.transformer('tiny-albert')
```

```python
xlnet.predict_proba(['bodoh punya orang'])
```

```python
tiny_albert.predict_proba(['danish busuk'])
```

```python
tiny_albert.predict_proba(['tengok la kau jalan kat mana'])
```

```python
xlnet.predict_proba(['aku nak tolong kau ni'])
```
